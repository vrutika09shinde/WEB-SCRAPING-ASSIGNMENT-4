# ANS =2

```python
import selenium
import pandas as pd
import time
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver = webdriver.Chrome()
```

```python
driver.get("https://www.bcci.tv/")
```

```python
series_title =[]

titles = driver.find_elements(By.XPATH,'//h5[@class="match-tournament-name ng-binding"]')
for i in titles:
    title = i.text
    series_title.append(title)

print(series_title)
```

```python
series_places =[]

titles = driver.find_elements(By.XPATH,'//div[@class="match-venue ng-scope"]')
for i in titles:
    title = i.text
    series_places.append(title)

print(series_places)
```

```python
series_date =[]

titles = driver.find_elements(By.XPATH,'//div[@class="match-dates ng-binding"]')
for i in titles:
    title = i.text
    series_date.append(title)

print(series_date)


series_time =[]

titles = driver.find_elements(By.XPATH,'//div[@class="match-time no-margin ng-binding"]')
for i in titles:
    title = i.text
    series_time.append(title)

print(series_time)
```
