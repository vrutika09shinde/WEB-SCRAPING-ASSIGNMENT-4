```python
driver = webdriver.Chrome()
```

```python
driver.get("https:/www.billboard.com/")
```

```python
song_name =[]

titles = driver.find_elements(By.XPATH,'//h3[@class="c-title  a-no-trucate a-font-primary-bold-s u-letter-spacing-0021 lrv-u-font-size-18@tablet lrv-u-font-size-16 u-line-height-125 u-line-height-normal@mobile-max a-truncate-ellipsis u-max-width-330 u-max-width-230@tablet-only"]')
for i in titles:
    title = i.text
    song_name.append(title)

print(song_name)
```
